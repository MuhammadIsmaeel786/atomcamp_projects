```python
msno.bar(df)
```

```python
values_list = list()
cols_list = list()
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())*100
    cols_list.append(col)
    values_list.append(pct_missing)
pct_missing_df = pd.DataFrame()
pct_missing_df['col'] = cols_list
pct_missing_df['pct_missing'] = values_list
```

```python
pct_missing_df
```

```python
pct_missing_df.loc[pct_missing_df.pct_missing > 0].plot(kind='bar', figsize=(12,8))
plt.show()
```

## Solve following question by using conditional statements (10)
##  How many studies have an enrollment greater than a certain threshold? 
##  How many clinical trials have 'No Results Available'? 
##  How many clinical trials are in an "Completed" and "Recruiting" status? 
##  How many clinical trials are related to 'COVID-19'? 
##  How many clinical trials started after January 1, 2020

```python
threshold = 10000
(df['Enrollment']<threshold).sum()
```

```python
(df['Study Results']=='No Results Available').sum()
```

```python
((df['Study Results']=='Completed') & (df['Status']=='Recruiting')).sum()
```

```python
(df['Acronym'] == "COVID-19").sum()
```

```python
(df['Start Date'] > '2020-02-01').sum()
```
