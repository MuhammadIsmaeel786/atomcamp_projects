### Q8. Use conditional statements and solve multiple questions a minimum of five.

**1 if City is islamabad and experince is less then 1 year**

```python
df[(df["City"]== 'Islamabad') & (df['Experience Required'] == "< 1 Year")]
```

**2 if City is islamabad and experince is less then 1 year and Job type fulltime**

```python
df[(df["City"]== 'Islamabad') & (df['Experience Required'] == "< 1 Year") & (df['Job Type'] == "Full Time Job")]
```

**3 if City is Karachi and experince is less then 1 year and Departement Marketing**

```python
df[(df["City"]== 'Karachi') & (df['Experience Required'] == "< 1 Year") & (df['Department'] == "Marketing Job")]
```

**4 if City is islamabad or lahore and Job type fulltime**

```python
df[(df["City"]== 'Islamabad') | (df["City"]== 'Lahore') & (df['Job Type'] == "Full Time Job")]
```

**5 if Company name is ibex,  Experience is Fresh Graduates**

```python
df[(df["Company Name"]== 'ibex, Pakistan') & (df['Experience Required'] == "Job for Fresh Graduates")]
```

### Q9  Apply the cut & qcut technique and compare the results.

My Dataset does't have the Integer values I am creating a small data set for Qcut and cut

```python
data =  [["Alice",10],["Bob",20],["Charlie",30],["David",40],["Eva",50],["Frank",60],["Grace",70],["Hannah",80],["Ivy",90],["Jack",100]]
```

```python
df1 = pd.DataFrame(data, columns=['Name', 'Age'])
df1
```

```python
pd.qcut(df1["Age"],q=4)
```

**Qcut**

```python
binname = ['young', 'mid young', 'middle ', 'experience']
df1['Age range'] = pd.qcut(df1["Age"],q=4,labels=binname)
```

```python
df1['Age range']
```

**Cut**

```python
pd.cut(df1["Age"],bins=4,labels = ['Teen','Young','Middle',"Old"])
```
